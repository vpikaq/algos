```python
import random
import usage_time
import matplotlib.pyplot as plt


def get_min(v: list):
    min_num = float('inf')
    for num in v:
        if num < min_num:
            min_num = num
    return min_num


N = 20 - 6

items = range(1, 10**5 * (20 - 6), 50000) 

func = usage_time.get_usage_time()(get_min)

times = [
    func([random.randint(1, 10) for _ in range(n)])
    for n in items
]

plt.plot(items, times, 'm-o', linewidth=1.5, markersize=5)
ax = plt.gca()

plt.title('time of getting min')
ax.set_xlabel('тumber of elements')
ax.set_ylabel('seconds')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import random, usage_time
import matplotlib.pyplot as plt


def matrix_mult(matrix_a: list, matrix_b: list):
    n = len(matrix_a)
    matrix_c = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            total = 0
            for k in range(n):
                total += matrix_a[i][k] * matrix_b[k][j]
            matrix_c[i][j] = total
    return matrix_c



N = 20 - 6
items = range(1, 10**2 * (20 - N), 100)
func = usage_time.get_usage_time()(matrix_mult)

times = [
    func(
        [
            [random.randint(1, 10) for _ in range(n)]
            for _ in range(n)
        ],
        [
            [random.randint(1, 10) for _ in range(n)]
            for _ in range(n)
        ]
    )
    for n in items
]

plt.plot(items, times, 'm-o', linewidth=1.5, markersize=4)
ax = plt.gca()

plt.title('time for matrix')
ax.set_xlabel('nxn')
ax.set_ylabel('seconds')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

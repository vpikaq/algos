```python
import random, usage_time
import matplotlib.pyplot as plt


def horner(v: list, x=2):
    result = 0
    for coeff in reversed(v):
        result = result * x + coeff
    return result

N = 20 - 6
items = range(1, 10000 * 14, 500 * 14)



func = usage_time.get_usage_time()(horner)


times = [
    func([random.randint(0, 100) for _ in range(n)]) 
    for n in items
]


plt.plot(items, times, 'm-o', linewidth=1.5, markersize=5)
ax = plt.gca()

plt.title('Execution time of Horner polynomial method')
ax.set_xlabel('number of elements')
ax.set_ylabel('seconds')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
